```python

import pandas as pd             # manipulation de données tabulaires
import sys
import os
import json                     # pour afficher proprement le JSON si besoin
import matplotlib.pyplot as plt
sys.path.append(os.path.abspath(".."))
```

```python
rank_probs = pd.read_parquet("../data/processed/monte_carlo_rank_probs.parquet")
```

```python
threshold=0.5
top7_prob = rank_probs.loc[:, 1:7].sum(axis=1)
teams_top7 = top7_prob.sort_values(ascending=False)
teams_top7 = teams_top7[teams_top7 > threshold].index.tolist()
bottom3_prob = rank_probs.loc[:, 16:18].sum(axis=1)
teams_bottom3 = bottom3_prob.sort_values(ascending=False)
teams_bottom3 = teams_bottom3[teams_bottom3 > threshold].index.tolist()

teams_other = rank_probs.index.difference(
    set(teams_top7).union(set(teams_bottom3))
).tolist()
```

```python
import matplotlib.pyplot as plt

teams_of_interest = teams_top7

plt.figure()

for team in teams_of_interest:
    plt.plot(
        rank_probs.columns, 
        rank_probs.loc[team],
        marker='o',
        label=team,
        linestyle='-'
    )

plt.xlabel("Classement")
plt.ylabel("Probabilité")
plt.title("Distribution de probabilité des classement finaux (après J31)")
plt.legend()
plt.grid()
plt.xlim(1, 14)
plt.ylim(0, 100)

plt.show()
```

```python
import matplotlib.pyplot as plt

teams_of_interest = teams_bottom3

plt.figure()

for team in teams_of_interest:
    plt.plot(
        rank_probs.columns, 
        rank_probs.loc[team],
        marker='o',
        label=team,
        linestyle='-'
    )

plt.xlabel("Classement")
plt.ylabel("Probabilité")
plt.title("Distribution de probabilité des classement finaux (après J31)")
plt.legend()
plt.grid()
plt.xlim(11, 18)
plt.ylim(0, 100)

plt.show()
```

```python
import matplotlib.pyplot as plt

teams_of_interest = teams_other

plt.figure()

for team in teams_of_interest:
    plt.plot(
        rank_probs.columns, 
        rank_probs.loc[team],
        marker='o',
        label=team,
        linestyle='-'
    )

plt.xlabel("Classement")
plt.ylabel("Probabilité")
plt.title("Distribution de probabilité des classement finaux (après J31)")
plt.legend()
plt.grid()
plt.xlim(1, 18)
plt.ylim(0, 60)

plt.show()
```
